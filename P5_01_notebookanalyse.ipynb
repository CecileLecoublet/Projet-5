#### 3.1.5 Imputation des valeurs manquantes <a class="anchor" id="section_3_1_5"></a>

Pour les variables où il y a des valeurs manquantes, on va les imputer par la moyenne.

```python
col = ["geolocation_lat", "geolocation_lng", "payment_sequential", "payment_installments", "payment_value"]
imp = SimpleImputer(missing_values=np.nan, strategy="mean", fill_value="missing")
df[col] = imp.fit_transform(df[col])
```

Pour les valeurs manquantes pour les dates, on va convertir les variables en datetime.<br>
Pour les valeurs manquantes de la variable "order_approved_at", qui est la date et l'heure de l'approbation de paiement. On prend la moyenne de la différence entre "order_approved_at" et "order_purchase_timestamp". Le temps en moyenne où la commande est achetée et l'approbation de paiement.<br>
Pour les valeurs manquantes de la variable "order_delivered_carrier_date", on fait la même chose que pour "order_approved_at". La moyenne de la différence entre l'heure de la validation de la commande et l'heure d'approbation du paiement.<br>
Pour les variables "order_estimated_delivery_date", "order_delivered_customer_date", on utilise la méthode "fillna" et on reporte les valeurs manquantes par rapport à l'autre colonne et inversement.

```python
# Conversion des variables au format datetime

time = ["order_purchase_timestamp", "order_approved_at",
        "order_delivered_carrier_date", "order_delivered_customer_date",
        "order_estimated_delivery_date"]

for cols in time:
    df[cols] = pd.to_datetime(df[cols],
                              format="%Y-%m-%d", errors="ignore")
```

```python
# Remplissage de df["order_approved_at"] avec df["order_purchase_timestamp"] plus un delta

time_mean = (df["order_approved_at"] - df["order_purchase_timestamp"]).mean()

df.reset_index(drop = True, inplace = True)
for i in range(0, len(df["order_approved_at"]), 1) :
    if pd.isna(df["order_approved_at"][i]) :
        df["order_approved_at"][i] = df["order_purchase_timestamp"][i] + time_mean
```

```python
# Remplissage de df["order_delivered_carrier_date"] avec df["order_approved_at"] plus un delta

time_mean = (df["order_delivered_carrier_date"] - df["order_approved_at"]).mean()

df.reset_index(drop = True, inplace = True)
for i in range(0, len(df["order_delivered_carrier_date"]), 1) :
    if pd.isna(df["order_delivered_carrier_date"][i]) :
        df["order_delivered_carrier_date"][i] = df["order_approved_at"][i] + time_mean
```

```python
# Remplissage de df[["order_estimated_delivery_date","order_delivered_customer_date"]] avec fillna

df[["order_estimated_delivery_date","order_delivered_customer_date"]] =\
                                df[["order_estimated_delivery_date","order_delivered_customer_date"]]\
                                    .fillna(method='ffill', axis = 1)
```

Les valeurs manquantes par rapport aux types de paiements, l'unique valeur manquante sera mise en "not_defined".

```python
# Comptage des valeurs

df["payment_type"].value_counts()
```

```python
df[df["payment_type"].isna()]
```

Pour les produits, qui ont des valeurs manquantes, on utilise la méthode "fillna" pour rapporter les produits écrits en portugais et on le reporte en anglais. Puis, on impute les valeurs avec les produits avec la méthode la plus fréquente.

```python
# Methode fillna()

df.reset_index(drop = True, inplace = True)

df[["product_category_name","product_category_name_english"]] =\
                                df[["product_category_name","product_category_name_english"]].fillna(method='ffill', axis = 1)

df.drop("product_category_name", axis = 1, inplace = True)
```

```python
# Imputation

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
```

```python
df = DataFrameImputer().fit_transform(df)
```
